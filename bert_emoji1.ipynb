```python
import csv

# Dictionary to store label counts
label_counts = {}

# Replace 'your_file.csv' with the path to your CSV file
file_path = '/content/drive/MyDrive/NlpInternship/constraint_Hindi_Train - Sheet1.csv'

# Assuming labels are in the first column (index 0)
column_index = 2

with open(file_path, newline='') as csvfile:
    reader = csv.reader(csvfile)
    next(reader)  # Skip header row if present

    for row in reader:
        labels = row[column_index].split(',')  # Split labels by comma (or your delimiter)
        for label in labels:
            label = label.strip()  # Remove leading/trailing spaces if any
            if label in label_counts:
                label_counts[label] += 1
            else:
                label_counts[label] = 1

print(label_counts)
```
